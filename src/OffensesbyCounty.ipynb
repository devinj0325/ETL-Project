```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csvfile = "../resources/ca_offenses_by_county.csv"
```

```python
offense_df = pd.read_csv(csvfile, encoding="ISO-8859-1")
```

```python
offense_df.head()
```

```python
#remove metro column
offense_df.drop("Metropolitan/Nonmetropolitan", axis=1)
offense_df
```

```python
#Extract only the 3 counties needed
counties_list = ['Los Angeles', 'Ventura', 'Orange']
county = offense_df[offense_df['County'].isin(counties_list)].copy()
```

```python
fips = [6037, 6059, 6111]
```

```python
county['FIPS'] = fips
```

```python
county.head()
```

```python
county.columns = ['county', 'metro/nonmetro','violent_crime', 'murder', 'rape','rape_legacy','robbery','assault','property_crime','burglary','larceny','vehicle_theft','arson', 'fips']
```

```python
county.head()
```

```python
#Set county as index
county.set_index('county')
```

```python
county.columns
```

```python
engine = create_engine('postgresql://postgres:postgres@localhost:5432/etl_project')
print(engine)
```

```python
# Use Pandas To Load CSV converted DataFrame into database
offense_df.to_sql(name='etl_project', con=engine, if_exists='append', index=True, index_label='id')
```

```python
# Check For Tables
engine.table_names()
```
